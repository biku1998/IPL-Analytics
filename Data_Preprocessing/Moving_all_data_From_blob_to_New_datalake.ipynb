```
connection_string  = None
container_name = None

with open(BLOB_CRED_JSON_PATH,'r') as F:
    info = json.load(F)
    connection_string = info['connection_string']
    container_name = info['container_name']
```

```
def list_blobs(connection_string,container_name):
    """
    This function will list out whatever files or folder are there in the container
    """
    try:
        blob_service_client = BlobServiceClient.from_connection_string(connection_string)

        container_client = blob_service_client.get_container_client(container_name)

        blob_list = container_client.list_blobs()

        for blob in blob_list:
            print("\t" + blob.name)

    except Exception as e:
        print('Exception :\n {}'.format(e))
```

```
def download_from_blob(connection_string,container_name,path_to_save_file = None):
    """
    This function will be used to download files from the blob
    """
    try:
        list_blobs(connection_string,container_name)
        print('Enter the file_name that you want to download, use the above list to choose files ...(copy file name with path)\n')

        file_name = input()

        blob_service_client = BlobServiceClient.from_connection_string(connection_string)

        container_client = blob_service_client.get_container_client(container_name) 

        blob_client = blob_service_client.get_blob_client(container=container_name, blob= file_name)

        download_file_path = None

        if '/' in file_name:
            file_name = file_name.split('/')[-1]


        if path_to_save_file is None:
            download_file_path = file_name
        else:
            download_file_path = path_to_save_file+file_name

        with open(download_file_path, "wb") as download_file:
            download_file.write(blob_client.download_blob().readall())
        
        if path_to_save_file is None:
            print('file downloaded in the local instance ..')
        else:
             print(f'file downloaded in {path_to_save_file} ..')

    except Exception as e:
        print('Exception :\n{}'.format(e))
```
