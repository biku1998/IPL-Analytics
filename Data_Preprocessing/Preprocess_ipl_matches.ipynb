```
DATA_SOURCE = "./ipl-analytics/raw/"
DATA_DEST = "./ipl-analytics/final/"
```

```
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_columns = 100
pd.options.display.max_rows = 500
```

```
df_matches = pd.read_csv(DATA_SOURCE+'ipl_matches.csv')
```

```
df_matches.head(10)
```

```
def remove_extra_char(toss_decision):
    if "elected to" in toss_decision:
        return toss_decision.split('to')[-1].strip()
    else:
        return toss_decision
```

```
df_matches['toss_decision']  = df_matches['toss_decision'].apply(remove_extra_char)
```

```
df_matches.sort_values('match_date',inplace=True)
```

```
df_matches.head()
```

```
df_matches['match_no'] = range(1,df_matches.shape[0]+1)
```

```
df_matches.to_csv(DATA_DEST+'ipl_matches.csv',index=False)
```
