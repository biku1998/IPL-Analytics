```
DATA_SOURCE = "./ipl-analytics/raw/"
DATA_DEST = "./ipl-analytics/final/"
```

```
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_columns = 100
pd.options.display.max_rows = 500
```

```
df_bowling = pd.read_csv(DATA_SOURCE+'ipl_bowling.csv')
```

```
df_bowling.head(10)
```

```
df_bowling['month'].unique()
```

```
month_dict = {'April':4, 'May':5, 'June':6, 'March':3, 'Apr':4, 'Jun':6}
```

```
df_bowling['month'] = df_bowling['month'].apply(lambda x : month_dict[x])
```

```
def create_date(col):
    if col['day'] != '27-28':
        return datetime.date(int(col['year']),int(col['month']),int(col['day']))
    else:
        return '---'
```

```
df_bowling['match_date'] = df_bowling.apply(create_date,axis = 1)
```

```
df_bowling.drop(['match_city','year','month','day'],axis=1,inplace=True)
```

```
df_bowling.head()
```

```
df_bowling.to_csv(DATA_DEST+'ipl_bowling.csv',index=False)
```
