```
DATA_SOURCE = "./ipl-analytics/raw/"
DATA_DEST = "./ipl-analytics/final/"
```

```
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_columns = 100
pd.options.display.max_rows = 500
```

```
df_batting = pd.read_csv(DATA_SOURCE+'ipl_batting.csv')
```

```
df_batting.head(100)
```

```
def filter_special_players(player):
    if "(c)" in player and "†" in player:
        return 'captain & wicket keeper'
    if '(c)' in player:
        return 'captain'
    if '†' in player:
        return 'wicket keeper'
    else:
        return 'normal player'
```

```
df_batting['special_role'] = df_batting['batsmen'].apply(filter_special_players)
```

```
# clean the batsmen i.e remove special symbols

def remove_special_char(player):
    if '(c)' in player and '†' in player:
        return " ".join(player.split(' ')[:-2])
        
    if '(c)' in player:
        return " ".join(player.split(' ')[:-1])
    if '†' in player:
        return " ".join(player.split(' ')[:-1])
    if '(c)' in player and '†' in player:
        return " ".join(player.split(' ')[:-2])
    else:
        return player
```

```
df_batting['batsmen'] = df_batting['batsmen'].apply(remove_special_char)
```

```
filter_ = df_batting['special_role'] != 'normal player'
df_batting[filter_].head()
```

```
df_batting['dismissal_mode'].unique()
```

```
def create_bowler_name(col):
    if col['dismissal_mode'] == 'caught' or col['dismissal_mode'] == 'stumped' or col['dismissal_mode'] == 'lbw' or col['dismissal_mode'] == 'hit wicket':
        return col['wicket_status'].split(' b ')[-1].strip() 
    elif col['dismissal_mode'] == 'bowled':
        return col['wicket_status'].split('b ')[-1].strip()
    else:
        return "no bowler"
```

```
df_batting['bowler_name'] = df_batting.apply(create_bowler_name,axis = 1)
```

```
df_batting.head()
```

```
def create_fielder_name(col):
    if col['dismissal_mode'] == 'caught':
        if '&' not in col['wicket_status']:
            return col['wicket_status'].split('c ')[-1].strip().split(' b ')[0].strip()
        else:
            return col['wicket_status'].split(' b ')[-1].strip()
    elif col['dismissal_mode'] == 'stumped':
        return col['wicket_status'].split('st ')[-1].strip().split(' b ')[0].strip()
    else:
        return "no fielder"
```

```
df_batting['fielder'] = df_batting.apply(create_fielder_name,axis = 1)
```

```
df_batting.head()
```

```
df_batting['is_fielder_involved'] = df_batting['fielder'].apply(lambda x : 1 if x != "no fielder" else 0)
```

```
df_batting.sample(10)
```

```
def create_date(col):
    if col['day'] != '27-28':
        return datetime.date(int(col['year']),int(col['month']),int(col['day']))
    else:
        return '---'
```

```
df_batting['match_date'] = df_batting.apply(create_date,axis = 1)
```

```
df_batting.drop(['match_city','year','month','day'],axis=1,inplace=True)
```

```
df_batting.sample(100)
```

```
df_batting.to_csv(DATA_DEST+'ipl_batting.csv',index=False)
```
