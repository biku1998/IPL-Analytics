```python
connection_string  = None
container_name = None

with open("/content/drive/My Drive/Classroom/project-archives/IPL Analysis/data/Azure_blob_info.json",'r') as F:
    info = json.load(F)
    connection_string = info['connection_string']
    container_name = info['container_name']

# print(connection_string)
# print(container_name)
```

```python
def list_blobs(connection_string,container_name):
  """
    This function will list out whatever files or folder are there in the container
  """
  try:
    blob_service_client = BlobServiceClient.from_connection_string(connection_string)
    container_client = blob_service_client.get_container_client(container_name)
    blob_list = container_client.list_blobs()
    for blob in blob_list:
      print("\t" + blob.name)
  except Exception as e:
    print('Exception :\n {}'.format(e))
```

```python
def store_to_blob(connection_string,container_name,path_to_csv,folder_name,file_name):
    """
    This function will store a df in azure blob storage,
    connection_string : connection_string for connecting to the blob
    container : container_name in which we have blob
    path_to_csv : csv file path , which you want to upload to blob
    folder_name : if there is a folder in container in which you want to store the file,
    file_name : what will be the name of the file in the blob
    """
    try:
        blob_service_client = BlobServiceClient.from_connection_string(connection_string)

        container_client = blob_service_client.get_container_client(container_name) 

        blob_client = blob_service_client.get_blob_client(container=container_name, blob= f"{folder_name}/{file_name}")

        with open(path_to_csv,'rb') as F:
            blob_client.upload_blob(F)

        print('File uploaded....')

    except Exception as e:
        print('Exception :\n{}'.format(e))
```

```python
store_to_blob(connection_string,container_name,
              path_to_csv=DATA_LAKE_FOLDER_PATH+'/team_captain_owner_information.csv',folder_name = 'raw',file_name = 'ipl_teams.csv')
```
