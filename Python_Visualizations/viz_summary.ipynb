```
import pandas as pd
import numpy as np
import plotly
import plotly.express as exp
import plotly.graph_objs as go
import plotly.figure_factory as ff
PATH = "/content/drive/My Drive/ipl-analytics/final/"
```

```
df = pd.read_csv(PATH+'ipl_summary.csv')
```

```
df.head()
```

```
# Let's see multi-level categorical feaures


trace_1 = go.Bar(x=df['Team'],y=df['Won'],marker = dict(color = '#3eca2b'),
		name = 'Won')

trace_2 = go.Bar(x=df['Team'],y=df['Lost'],marker = dict(color = '#c84a2c'),
		name = 'Lose')

# trace_3 = go.Bar(x=df['NOC'],y=df['Bronze'],marker = dict(color = '#CD7F32'),
# 		name = 'Bronze')


data = [trace_1,trace_2]


layout = go.Layout(title = "IPL Teams Win and Lose all seasons")

fig = go.Figure(data = data,layout = layout)

fig.show()
```
